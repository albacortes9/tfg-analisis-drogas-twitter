```python
import pandas as pd
import matplotlib
matplotlib.rcParams['font.family'] = 'DejaVu Sans'
import matplotlib.pyplot as plt
```

```python
data = {
    'Método': ['DT', 'RF', 'AdaBoost', 'Bagging', 'LR', 'XGBoost', 'MLP'],
    'F1': [0.744557, 0.839184, 0.786910, 0.842949, 0.575022, 0.885246, 0.872434]
}

df_f1 = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plt.bar(df_f1['Método'], df_f1['F1'], color='skyblue')
plt.xlabel('Modelo')
plt.ylabel('F1-Score')
plt.title('Comparación de F1-Score en Test (Cost-Sensitive)')
plt.ylim(0, 1)
plt.show()
```

```python
# Datos de F1-Score en CV y Test para cada modelo (extraídos de tus tablas)
modelos = ['DT', 'RF', 'AdaBoost', 'Bagging', 'LR', 'XGBoost', 'MLP']
f1_cv = [0.970, 0.993, 0.949, 0.996, 0.953, 0.990, 0.995]  # F1-Score CV (promedio de la tabla 1)
f1_test = [0.685440, 0.849822, 0.597968, 0.828017, 0.610536, 0.872093, 0.843816]  # F1-Score Test (tabla 2)

plt.figure(figsize=(10, 6))
plt.plot(modelos, f1_cv, marker='o', label='F1-Score CV')
plt.plot(modelos, f1_test, marker='o', label='F1-Score Test')
plt.fill_between(modelos, f1_cv, f1_test, color='gray', alpha=0.2, label='Brecha')

plt.xlabel('Modelo')
plt.ylabel('F1-Score')
plt.title('Comparación de F1-Score en CV vs Test (SMOTE_PRE_CV)')
plt.ylim(0, 1.05)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# Datos de F1-Score y AUPRC en Test para cada modelo (SMOTE_IN_CV)
modelos_smote = ['DT', 'RF', 'AdaBoost', 'Bagging', 'LR', 'XGBoost', 'MLP']
f1_test_smote = [0.584521, 0.855714, 0.497588, 0.787037, 0.454800, 0.878832, 0.850000]
auprc_test_smote = [0.486030, 0.919389, 0.843038, 0.898005, 0.853682, 0.932542, 0.918360]

x = range(len(modelos_smote))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar([i - width/2 for i in x], f1_test_smote, width=width, label='F1-Score', color='skyblue')
plt.bar([i + width/2 for i in x], auprc_test_smote, width=width, label='AUPRC', color='orange')
plt.xticks(x, modelos_smote)
plt.xlabel('Modelo')
plt.ylabel('Score')
plt.title('Comparación de F1-Score y AUPRC en Test (SMOTE In-CV)')
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
# Datos de F1-Score y AUPRC en Test para la estrategia "Sin Balanceo"
modelos_sin_balanceo = ['DT', 'RF', 'AdaBoost', 'Bagging', 'LR', 'XGBoost', 'MLP']
f1_test_sin_balanceo = [0.740845, 0.842962, 0.786910, 0.837359, 0.815789, 0.880734, 0.853550]
auprc_test_sin_balanceo = [0.561824, 0.918662, 0.847976, 0.893794, 0.882524, 0.930942, 0.921938]

x = range(len(modelos_sin_balanceo))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar([i - width/2 for i in x], f1_test_sin_balanceo, width=width, label='F1-Score', color='skyblue')
plt.bar([i + width/2 for i in x], auprc_test_sin_balanceo, width=width, label='AUPRC', color='orange')
plt.xticks(x, modelos_sin_balanceo)
plt.xlabel('Modelo')
plt.ylabel('Score')
plt.title('Comparación de F1-Score y AUPRC en Test (Sin Balanceo)')
plt.ylim(0, 1)
plt.legend()
plt.show()
```
